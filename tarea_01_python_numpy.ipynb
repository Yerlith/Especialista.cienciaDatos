**P7.** Dado el arreglo `a = np.arange(1, 13)`, usa `reshape` para convertirlo en una matriz de 3 × 4.  
Luego imprime el valor máximo, el valor mínimo y la posición (índice) del máximo.

```python
a = np.arange(1, 13)

# Tu respuesta aquí
matriz = a.reshape(3, 4)
maximo = matriz.max()
minimo = matriz.min()
pos_max = np.argmax(matriz)

print("\nValor máximo:", maximo)
print("Valor mínimo:", minimo)
print("Posición (índice) del máximo:", pos_max)

```

**P8.** Crea dos arreglos de 5 elementos con `np.random.randint` (valores entre 1 y 20).

**P9.** Dado el arreglo `notas = np.array([8.5, 6.0, 9.2, 4.3, 7.8, 5.5, 10.0, 3.1])`,
