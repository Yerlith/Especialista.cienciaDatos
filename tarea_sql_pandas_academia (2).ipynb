## 3. Cargar tablas a SQLite

### Punto 4. Agregación simple

Calcula:
- número total de cursos
- precio promedio
- precio mínimo
- precio máximo

Usa la tabla `cursos`.

**Entregable:**
- SQL
- pandas

```python
# Punto 4 - SQL
# Escribe aquí tu consulta
query = """
SELECT
    COUNT(curso_id) AS total_cursos,
    AVG(precio) AS precio_promedio,
    MIN(precio) AS precio_minimo,
    MAX(precio) AS precio_maximo
FROM cursos;
"""
run_sql(query)
```

```python
# Punto 4 - pandas
# Escribe aquí tu solución
cursos.agg(
    total_cursos=('curso_id', 'count'),
    precio_promedio=('precio', 'mean'),
    precio_minimo=('precio', 'min'),
    precio_maximo=('precio', 'max')
)
```

```python
# Punto 5 - SQL
# Escribe aquí tu consulta
query = """
SELECT
    categoria,
    COUNT(curso_id) AS cantidad_cursos
FROM cursos
GROUP BY categoria
ORDER BY cantidad_cursos DESC;
"""
run_sql(query)
```

### Punto 6. JOIN entre tablas

Muestra:
- nombre del estudiante
- nombre del curso
- estado de la inscripción

Usa `estudiantes`, `inscripciones` y `cursos`.

**Entregable:**
- SQL
- pandas

```python
# Punto 6 - SQL
# Escribe aquí tu consulta
query = """
SELECT
    e.nombre AS nombre_estudiante,
    c.nombre_curso,
    i.estado AS estado_inscripcion
FROM estudiantes e
JOIN inscripciones i ON e.estudiante_id = i.estudiante_id
JOIN cursos c ON i.curso_id = c.curso_id;
"""
run_sql(query)
```

```python
# Punto 6 - pandas
# Escribe aquí tu solución
merged_df = pd.merge(estudiantes, inscripciones, on='estudiante_id')
merged_df = pd.merge(merged_df, cursos, on='curso_id')
merged_df[['nombre', 'nombre_curso', 'estado']]
```

### Punto 7. LEFT JOIN

Trae todos los estudiantes, incluso si no tuvieran inscripciones, mostrando:
- `estudiante_id`
- `nombre`
- `inscripcion_id`

**Entregable:**

```python
# Punto 7 - SQL
# Escribe aquí tu consulta
query = """
SELECT
    e.estudiante_id,
    e.nombre,
    i.inscripcion_id
FROM estudiantes e
LEFT JOIN inscripciones i ON e.estudiante_id = i.estudiante_id;
"""
run_sql(query)
```

### Punto 8. HAVING

Calcula cuántas inscripciones tiene cada estudiante y muestra solo aquellos con **2 o más**.

Debes mostrar:
- `estudiante_id`
- `nombre`
- `n_inscripciones`

**Entregable:**
- SQL
- pandas

```python
# Punto 8 - SQL
# Escribe aquí tu consulta
query = """
SELECT
    e.estudiante_id,
    e.nombre,
    COUNT(i.inscripcion_id) AS n_inscripciones
FROM estudiantes e
JOIN inscripciones i ON e.estudiante_id = i.estudiante_id
GROUP BY e.estudiante_id, e.nombre
HAVING COUNT(i.inscripcion_id) >= 2;
"""
run_sql(query)
```

```python
# Punto 8 - pandas
# Escribe aquí tu solución
merged_df = pd.merge(estudiantes, inscripciones, on='estudiante_id', how='inner')
inscripciones_por_estudiante = merged_df.groupby(['estudiante_id', 'nombre']).size().reset_index(name='n_inscripciones')
inscripciones_por_estudiante[inscripciones_por_estudiante['n_inscripciones'] >= 2]
```

### Punto 9. CASE WHEN / segmentación

Clasifica los cursos según su precio:
- `Bajo` si el precio es menor a 120
- `Medio` si el precio está entre 120 y 180 inclusive
- `Alto` si el precio es mayor a 180

Debes mostrar:
- `nombre_curso`
- `precio`
- `segmento_precio`

**Entregable:**
- SQL
- pandas

```python
# Punto 9 - SQL
# Escribe aquí tu consulta
query = """
SELECT
    nombre_curso,
    precio,
    CASE
        WHEN precio < 120 THEN 'Bajo'
        WHEN precio >= 120 AND precio <= 180 THEN 'Medio'
        ELSE 'Alto'
    END AS segmento_precio
FROM cursos;
"""
run_sql(query)
```

```python
# Punto 9 - pandas
# Escribe aquí tu solución
def clasificar_precio(precio):
    if precio < 120:
        return 'Bajo'
    elif 120 <= precio <= 180:
        return 'Medio'
    else:
        return 'Alto'

cursos['segmento_precio'] = cursos['precio'].apply(clasificar_precio)
cursos[['nombre_curso', 'precio', 'segmento_precio']]
```

### Punto 10. Ejercicio final integrador

Calcula el total pagado por categoría de curso.

Debes:
- unir `cursos`, `inscripciones` y `pagos`
- agrupar por `categoria`
- sumar `monto`
- ordenar de mayor a menor

**Entregable:**
- SQL
- pandas

```python
# Punto 10 - SQL
# Escribe aquí tu consulta
query = """
SELECT
    c.categoria,
    SUM(p.monto) AS total_pagado
FROM cursos c
JOIN inscripciones i ON c.curso_id = i.curso_id
JOIN pagos p ON i.inscripcion_id = p.inscripcion_id
GROUP BY c.categoria
ORDER BY total_pagado DESC;
"""
run_sql(query)
```

```python
# Punto 10 - pandas
# Escribe aquí tu solución
merged_df = pd.merge(cursos, inscripciones, on='curso_id')
merged_df = pd.merge(merged_df, pagos, on='inscripcion_id')
total_pagado_por_categoria = merged_df.groupby('categoria')['monto'].sum().reset_index(name='total_pagado')
total_pagado_por_categoria.sort_values(by='total_pagado', ascending=False)
```
